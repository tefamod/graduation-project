```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
#from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score
#from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report,confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Assuming your data is in a CSV file named 'your_data.csv'
data = pd.read_csv('Robot_train_4s _ without_y_one_joint_int.csv')

# 1. Data Preprocessing
# Example: Handle missing values
data = data.dropna()
```

```python
print(data.keys())
```

```python
y = data.pop('Label')
x = data
print(x.shape,y.shape)
```

```python
x
```

```python
x =  data[['X', 'Z', 'slop(x)', 'slop(x/z)', 'slop(z)']]
X_train,X_test,y_train,y_test= train_test_split( x,y, test_size=0.2, random_state=42)
```

```python
x
```

```python
rf_model_5 = RandomForestClassifier(random_state=42,max_depth=7,n_estimators=4)
rf_model_5.fit(X_train, y_train)

# 4. Model Evaluation
y_pred_5 = rf_model_5.predict(X_test)
accuracy_5 = accuracy_score(y_test,y_pred_5)
#print(f'Random Forest Accuracy: {accuracy_5}')
print("5 Features : 'X', 'Z', 'slop(x)', 'slop(x/z)', 'slop(z)'")
#cm=confusion_matrix(y_test,y_pred_7)
#plt.figure(figsize =(6,6))
#sns.heatmap(cm,annot=True,cmap='Blues',fmt='g',xticklabels=list(set(y_test)),yticklabels=list(set(y_test)))
#plt.xlabel('Predicted Label')
#plt.ylabel('True label')
#plt.show()
print(classification_report(y_test,y_pred_5))
print('Accuracy : ', accuracy_score(y_test, y_pred_5))
```
